### Analysis

Following are some key observations that can be inferred based on the data analysis:

• The average tumor volume for mice treated with Capomulin (40.676 mm3) and Ramicane (40.217 mm3) was lowest as compared to the average tumor volume of treatments with other drugs (ranging between 52.321 mm3 and 55.236 mm3). The tumor volume data of both Capomulin and Ramicane also had the lowest standard deviation, variance and standard error as compared with tumor volumes of mice treated with other drugs.

• Based on the boxplot, the upper and lower boundaries of the tumor volume data at the last timepoint for mice treated with Capomulin and Ramicane were lower compared to the data for Infubinol and Ceftamin. The box plot also shows a relatively lower average tumor volume for mice treated with Capomulin and Ramicane. This indicates that these two drug regimens were the most effective in reducing the tumor size by the end of the treatment period.

• The line chart shows an interesting trend for the mouse ‘l509’ treated with Capomulin. The tumor volume of the mouse treated with this drug increased gradually from 45 mm3 to 48 mm3 in the first 20 days. The tumor volume then dropped sharply till day 25 of treatment back to about 45 mm3. It then increased slightly until day 30. There was a sharp decline from day 30 to 35 where the tumor volume reached about 40 mm3. After that, the tumor volume started to increase again slightly until day 45. This trend in increase and decrease of tumor volume during the treatment period shows an interesting relationship. It also indicates that the effectiveness of drug might not consistently decrease the tumor volume and it could in fact increase slightly during the treatment period. However, the end result does prove the effectiveness of the drug.

• The scatter plot and regression line indicate a linear relationship between the average tumor volume and average weight of the mice treated with Capomulin. The correlation coefficient of 0.84 indicates a strong positive correlation. This implies that the mice with greater weight tend to show a weaker response to drugs tested as they have tumors with bigger volumes.

• Overall, all the different analyses show consistent findings and indicate Capomulin and Ramicane to be the most effective drug regimens. Out of the four top drugs analyzed, only Infubinol has one outlier which affirms the reliability of the data.

```python
# Dependencies and Setup
# Matplotlib inline allows rendering of Matplotlib plots
%matplotlib inline

# Import pyplot to render graphing tools
import matplotlib.pyplot as plt

# import pandas for data manipulation and analysis
import pandas as pd

# Import scipy.stats for statistical functions
import scipy.stats as st
from scipy.stats import linregress

# Import numpy for working with numerical Data
import numpy as np
```

```python
# Prep Data
# Files for study & where to find them

mouse_meta_data_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the data and study results
# pd.read takes .cvs file and reads data into a DataFrame object
mouse_metadata = pd.read_csv(mouse_meta_data_path)
study_results = pd.read_csv(study_results_path)

# Rquirenent 1 Datasets are merged intno a single dataframe 6 points
# Merge into a single dataset
# pd.mergre combines multiple data frames into one object
data_df = pd.merge(study_results, mouse_metadata, how = 'outer', on = 'Mouse ID')

# Print data preview to verify 
data_df.head() 
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# # Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# # mean, median, variance, standard deviation, and SEM of the tumor volume. 

tumor_data = cleaned_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)']
tumor_volume_mean = tumor_data.mean()
tumor_volume_median = tumor_data.median()
tumor_volume_variance = tumor_data.var()
tumor_voluem_std = tumor_data.std()
tumor_volume_sem = tumor_data.sem()

# # Assemble the resulting series into a single summary DataFrame.
# DataFrame Drug Regimen/Mean/Median/Variance/Std Dev/SEm 

tumor_summary = pd.DataFrame({"Mean Tumor Volume" : tumor_volume_mean,
                              "Median Tumor Volume" : tumor_volume_median,
                              "Tumor Volume Variance" : tumor_volume_variance,
                              "Tumor Volume Std Dev." : tumor_voluem_std,
                              "Tumor Volume SEM" : tumor_volume_sem})
tumor_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
# aggg function is used to apply multiple functions to a column in a DataFrame


tumor_data_summary = cleaned_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg(['mean' , 'median' , 'var', 'std' , 'sem'])
tumor_data_summary
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
group_df=cleaned_df.groupby(cleaned_df['Mouse ID'])['Timepoint'].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
treatments_df=pd.merge(cleaned_df, group_df, on=('Mouse ID', 'Timepoint'), how='right')
treatments_df.reset_index(drop=True)
treatments_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin' , 'Ramicane' , 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers
# Locate the rows which contain mice on each drug and get the tumor volumes

for drug in treatments:
    drug_data=treatments_df.loc[treatments_df["Drug Regimen"] == drug]
    tumor_vol_data.append(drug_data["Tumor Volume (mm3)"])

    # Determine outliers using upper & lower bounds
    quartiles = drug_data["Tumor Volume (mm3)"].quantile([0.25 , 0.5, 0.75])
    lowerq=quartiles[0.25]
    upperq=quartiles[0.75]
    iqr = upperq - lowerq

    lower_bound=round(lowerq - (1.5*iqr) ,5)
    upper_bound=round(upperq + (1.5*iqr),5)

    outliers = drug_data.loc[(drug_data["Tumor Volume (mm3)"]>=upper_bound)|
                              (drug_data['Tumor Volume (mm3)']<=lower_bound)]
    outliers_values = outliers["Tumor Volume (mm3)"]

    print(f"{drug},has lower boundary of {lower_bound} and upper boudary of {upper_bound}")
    print(f"{drug}'s potential outliers: {outliers_values}")
    print(f"_____________________________________________________________________________________")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

# Specify display details for the outliers
marker_details = dict(markerfacecolor='r', markersize=14, markeredgecolor='black')

# Create chart and add title and labels
boxplot, ax1 = plt.subplots()
ax1.boxplot(tumor_vol_data, flierprops=marker_details)

plt.title("Tumor Value for different Treatments")
plt.xlabel('Name of Treatment')
plt.ylabel("Final Tumor Volume (mm3)")
plt.rcParams["figure.figsize"]=(7,5)
plt.rcParams["font.size"]=10
plt.tick_params(labelsize=10)
plt.xticks([1, 2, 3, 4], treatments)

plt.show()
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
line_data = cleaned_df.loc[cleaned_df["Mouse ID"]=="b128"]
timepoint = line_data["Timepoint"]
tumor_volume = line_data["Tumor Volume (mm3)"]

# Plot the data
line1 = plt.plot(timepoint, tumor_volume)


# Add chart title and labels
plt.title("Capomulin treatment of mouse b128")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume")

# Display Chart
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# Get the data for Average Tumor Volume and Mouse Weight for Capomulin and group by Mouse ID
scatter_data = cleaned_df.loc[cleaned_df["Drug Regimen"] == "Capomulin"]
scatter_average_data=scatter_data.groupby("Mouse ID").mean(numeric_only=True)
avg_tumor_volume = scatter_average_data["Tumor Volume (mm3)"]
mouse_weight = scatter_average_data["Weight (g)"]

# print(scatter_average_data.head())

# Plot the data
scatter1 = plt.scatter(mouse_weight, avg_tumor_volume)

# PLot
plt.title('Average Tumor Volume vs. Average Mouse Weight')
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

# Show the chart
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Calculate the correlation coefficient
correlation = st.pearsonr(mouse_weight, avg_tumor_volume)
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")

# Add the linear regression equation and line to the scatter plot
# Re-name data arrays to x_values and y_values for easy reference
x_values = mouse_weight
y_values = avg_tumor_volume

# Calculate the values for linear regression model and draft the linear regression equation
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Chart the data into scatter plot
figure8 = plt.scatter(x_values,y_values)

# Draw the slope and insert the linear equation
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(19,36),fontsize=12,color="red")

# Add chart title and labels
plt.title("Average Tumor Volume vs. Average Mouse Weight")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

# Show the chart
plt.show()
```
